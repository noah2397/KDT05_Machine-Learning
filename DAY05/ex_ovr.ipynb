```python
#OVR 모듈 활용

from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np

df=pd.read_csv("../data/fish.csv")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[df.columns[1:]], df['Species'], test_size=0.3, random_state=4)
```

```python
model=LogisticRegression(max_iter=10000, solver="liblinear")
```

```python
ovoModel=OneVsOneClassifier(model)
ovoModel.fit(x_train, y_train)
```

```python
print(ovoModel.classes_)
print(ovoModel.feature_names_in_)
print(len(ovoModel.estimators_)) # 6*7/2=21개 
```

```python
print(f"train score : {ovoModel.score(x_train, y_train)}\n test score : {ovoModel.score(x_test, y_test)}")
```

```python
y_pre=ovoModel.predict(x_test)
a=pd.DataFrame(np.round(ovoModel.decision_function(x_test),2), columns=ovoModel.classes_)
b=pd.DataFrame(y_pre, columns=["Pre_Y"])
pd.concat([a,b], axis=1)
```

```python
# OvR 기반 학습 진행
ovrModel=OneVsRestClassifier(model)
ovrModel.fit(x_train, y_train)
```

```python
print(ovrModel.classes_)
print(ovrModel.feature_names_in_)
print(len(ovrModel.estimators_)) # 7개 
```

```python
ovrModel.score(x_train, y_train), ovrModel.score(x_test, y_test)
```
